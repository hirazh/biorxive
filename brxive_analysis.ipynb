```python
# DataFrames
df1 = pd.read_csv(r'C:\Users\Hiras\Analysis_Hiras\biorxive\output\brxive_df20.csv')
df1 = df1.drop('Unnamed: 0', axis=1)

df2 = pd.read_csv(r'C:\Users\Hiras\Analysis_Hiras\biorxive\output\brxive_df21.csv')
df2 = df2.drop('Unnamed: 0', axis=1)

df3 = pd.read_csv(r'C:\Users\Hiras\Analysis_Hiras\biorxive\output\brxive_df22.csv')
df3 = df3.drop('Unnamed: 0', axis=1)

df4 = pd.read_csv(r'C:\Users\Hiras\Analysis_Hiras\biorxive\output\brxive_df23.csv')
df4 = df4.drop('Unnamed: 0', axis=1)
```
