## K Means Clustering
Clustering the catogeries to indentify the sub catogeries in them. For an example to identify the main clusters in the Bioinformatics category

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Creating a different dataset for the Bioinformatics category
binformatics_df = brxive_df[brxive_df['preprint_category'] == 'bioinformatics']
binformatics_list = binformatics_df['preprint_abstract'].to_list()
# binformatics_list
```

```python
len(binformatics_list)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
import string

nlp = spacy.load("en_core_web_sm")

punctuations = string.punctuation #list of punctuation to remove from text
stopwords = list(STOP_WORDS)

def spacy_tokenizer(sentence):
    ''' Function to preprocess text of scientific papers 
        (e.g Removing Stopword and puntuations)'''
    mytokens = nlp(sentence)
    # transform to lowercase and then split the scentence
    mytokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens ]
    #remove stopsword an punctuation
    mytokens = [ word for word in mytokens if word not in stopwords and word not in punctuations ] 
    mytokens = " ".join([i for i in mytokens]) 
    return mytokens
```

```python
from tqdm import tqdm
from IPython.utils import io
```

```python
binformatics_list_tknzd = []
for item in tqdm(binformatics_list, desc="Tokenizing the list"):
    binformatics_list_tknzd.append(spacy_tokenizer(item))
```

```python
# create the vectorizer
vectorizer = TfidfVectorizer()

# fit the vectorizer to the data and transform the data
X = vectorizer.fit_transform(binformatics_list_tknzd)

# convert the sparse matrix to a dense array
X = X.toarray()

# apply Spacy's Doc objects to obtain dense vector representations of each document
nlp = spacy.load("en_core_web_sm")
X = np.array([nlp(doc).vector for doc in binformatics_list_tknzd])
```

```python
X
```

```python
# create a list of WCSS values for different k values
wcss = []
for k in range(1, 51):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# plot the elbow curve
plt.plot(range(1, 51), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# apply K-Means clustering
k = 10 # number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(X)
```

```python
# plot the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="Set2", s=100, alpha=0.7)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Clusters of Text Data")
plt.show()
```
